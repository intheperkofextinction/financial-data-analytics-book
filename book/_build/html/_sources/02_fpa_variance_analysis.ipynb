## 1. What is Financial Planning & Analysis (FP&A)?


#### Core Definition
Financial Planning & Analysis (FP&A) is a finance function that focuses on
planning future financial performance and analyzing actual results
to support management decision-making.

Unlike accounting, which records historical transactions,
FP&A is forward-looking and analytical in nature.


FP&A typically sits between finance, operations, and senior management.

Its role is to translate financial data into insights that help leaders:
- Allocate resources efficiently
- Understand financial risks and opportunities
- Evaluate business performance against plans


#### FP&A vs Accounting vs Data Analytics
FP&A typically sits between finance, operations, and senior management.

Its role is to translate financial data into insights that help leaders:
- Allocate resources efficiently
- Understand financial risks and opportunities
- Evaluate business performance against plans


While these functions work closely together, their objectives differ:

- **Accounting** focuses on accuracy, compliance, and historical reporting.
- **FP&A** focuses on planning, forecasting, and performance analysis.
- **Data Analytics** focuses on extracting insights from data using statistical and technical tools.

A Financial Data Analyst in an FP&A role combines
financial knowledge with data analysis techniques.


#### Core Responsibilities of FP&A
Typical FP&A responsibilities include:
- Budget preparation and management
- Financial forecasting and reforecasting
- Budget vs Actual analysis
- Variance analysis
- Management reporting and decision support


FP&A plays a critical role in aligning financial performance
with business strategy.

By identifying deviations from plans early,
FP&A enables organizations to take corrective actions
before financial issues escalate.


## 2. Budgeting Basics



#### What is a Budget?
A budget is a financial plan that outlines expected revenues,
costs, and expenses over a specific period of time.

Budgets are prepared before a period begins and serve as a
benchmark against which actual performance is measured.


The primary purposes of budgeting include:
- Planning future financial performance
- Allocating resources efficiently
- Setting financial targets
- Controlling costs
- Measuring performance against expectations


**Types of Budgets**

Organizations typically prepare multiple types of budgets,
depending on their size and complexity.

Common budget types include:
- **Operating Budget**: Covers day-to-day revenues and expenses
- **Capital Budget**: Covers long-term investments such as equipment or infrastructure


**Revenue and Expense Budgeting**

Budgeting usually separates revenues and expenses.

- Revenue budgets estimate future sales based on assumptions such as pricing, volume, and market conditions.
- Expense budgets estimate costs required to generate those revenues, including fixed and variable costs.


Budgets are based on assumptions rather than certainty.

Common assumptions include:
- Expected growth rates
- Cost inflation
- Market demand
- Operational capacity

The quality of a budget depends heavily on the quality of its assumptions.


**Budgeting Assumptions**

Budgets are prepared for different time horizons:
- Annual budgets for long-term planning
- Quarterly budgets for performance monitoring
- Monthly budgets for operational control


## 3. Forecasting Fundamentals


**What is a Forecast?**

A forecast is an updated estimate of future financial performance
based on the most recent available information.

Unlike budgets, forecasts are revised periodically
to reflect changes in business conditions.


**Budget vs Forecast**

Although both budgets and forecasts deal with future performance,
they serve different purposes.

- **Budget**: A fixed financial plan prepared before the period begins
- **Forecast**: A dynamic estimate that changes as new information becomes available


**Why Forecasts Are Updated**

Forecasts are updated to account for:
- Changes in market demand
- Pricing adjustments
- Cost fluctuations
- Operational constraints
- Unexpected events

Regular forecasting allows management to react
before issues significantly impact results.


**Types of Forecasting Approaches**

Types of Forecasting Approaches
Common forecasting approaches include:

- **Top-down forecasting**: Targets are set by senior management
- **Bottom-up forecasting**: Estimates are built by individual departments
- **Driver-based forecasting**: Forecasts are linked to key business drivers such as volume or pricing

**Rolling Forecasts**

A rolling forecast continuously extends the forecast horizon
by adding a new period as time progresses.

For example, a 12-month rolling forecast always looks
12 months ahead, regardless of the current month.


**Forecasting in FP&A**

In FP&A, forecasting is not about precise prediction.
It is about maintaining visibility into future performance
and enabling informed decision-making.


## 4. Budget vs Actual Analysis


**What Are “Actuals”?**

Actuals represent the real financial results recorded
during a specific period.

They are derived from accounting systems
after transactions have occurred.


**Purpose of Budget vs Actual Analysis**

Budget vs Actual analysis compares planned financial performance
with actual results.

This analysis helps organizations:
- Identify deviations from plans
- Understand reasons for overperformance or underperformance
- Improve future planning and forecasting


**Common Dimensions of Comparison**

Budget vs Actual analysis is typically performed across:
- Time periods (monthly, quarterly, yearly)
- Business units or departments
- Revenue and cost categories


Example Budget vs Actual Data

Consider a simplified monthly income statement
for a hypothetical company.


````python
import pandas as pd

data = {
    "Month": ["January", "February", "March"],
    "Budget_Revenue": [100000, 120000, 130000],
    "Actual_Revenue": [95000, 125000, 128000],
    "Budget_Expense": [60000, 70000, 75000],
    "Actual_Expense": [62000, 68000, 77000],
}

df = pd.DataFrame(data)
df
````

The table above shows planned and actual revenues
and expenses for three months.

At this stage, no conclusions are drawn.
The purpose is to establish a structured dataset
that enables comparison and analysis.


**Why Structure Matters**

Budget vs Actual analysis requires data
to be organized in a consistent and comparable format.

Well-structured data simplifies:
- Variance calculation
- Trend analysis
- Performance reporting


## 5. Variance Analysis


**What is Variance?**

Variance represents the difference between planned values
(such as budgets or forecasts) and actual results.

It is a key concept in FP&A because it highlights
where performance deviates from expectations.



**Types of Variance**

Variance can be measured in two primary ways:

- **Absolute Variance**: The numerical difference between actual and budgeted values
- **Percentage Variance**: The relative difference expressed as a percentage


**Favorable vs Unfavorable Variance**

Variances are commonly classified as:

- **Favorable Variance (F)**: Actual performance is better than planned
- **Unfavorable Variance (U)**: Actual performance is worse than planned

The interpretation depends on the metric being analyzed.


**Interpreting Revenue and Cost Variances**

For revenues:
- Actual revenue higher than budget is generally favorable
- Actual revenue lower than budget is generally unfavorable

For costs:
- Actual costs lower than budget are generally favorable
- Actual costs higher than budget are generally unfavorable


**Why Variance Exists**

Variances occur due to multiple factors, including:
- Changes in demand
- Pricing differences
- Cost inflation
- Operational inefficiencies
- External economic conditions


**Analytical Purpose of Variance Analysis**

The objective of variance analysis is not to assign blame.

Its purpose is to:
- Identify performance drivers
- Improve future budgeting and forecasting
- Support informed management decisions


**Preparing for Quantitative Analysis**

To move from observation to insight,
variances must be calculated systematically.

In the next section, variance analysis will be implemented
using Python to quantify performance differences.


## 6. Variance Analysis Using Python


**Objective of This Section**

In this section, variance analysis is implemented using Python.

The goal is to calculate revenue and expense variances
and interpret the results in a structured manner.


**Calculating Revenue Variance**

Revenue variance is calculated as the difference
between actual revenue and budgeted revenue.


````python
df["Revenue_Variance"] = df["Actual_Revenue"] - df["Budget_Revenue"]
df
````

**Interpreting Revenue Variance**

A positive revenue variance indicates that actual revenue
exceeded the budget, which is generally favorable.

A negative revenue variance indicates underperformance
relative to the plan.

**Calculating Expense Variance**

Expense variance is calculated as the difference
between actual expenses and budgeted expenses.


````python
df["Expense_Variance"] = df["Actual_Expense"] - df["Budget_Expense"]
df
````

**Interpreting Expense Variance**

For expenses, interpretation differs from revenue.

A positive expense variance indicates higher-than-planned costs,
which is generally unfavorable.

A negative expense variance indicates cost savings
relative to the budget.


**Percentage Variance**

Percentage variance helps normalize differences
and makes comparisons across periods easier.


````python
df["Revenue_Variance_%"] = (df["Revenue_Variance"] / df["Budget_Revenue"]) * 100
df["Expense_Variance_%"] = (df["Expense_Variance"] / df["Budget_Expense"]) * 100

df
````

**Why Percentage Variance Matters**

Absolute variances show the magnitude of deviation,
while percentage variances show relative impact.

FP&A analysts use both views to prioritize issues
and communicate insights effectively.


**Clean Summary Table**

For reporting purposes, analysts often prepare
a simplified variance summary table.

````python
summary_df = df[
    [
        "Month",
        "Budget_Revenue",
        "Actual_Revenue",
        "Revenue_Variance",
        "Revenue_Variance_%",
        "Budget_Expense",
        "Actual_Expense",
        "Expense_Variance",
        "Expense_Variance_%"
    ]
]

summary_df
````

## 7. Variance Analysis Using SQL


**Why SQL Matters in FP&A**

In many organizations, financial data is stored in databases
rather than spreadsheets or notebooks.

FP&A analysts frequently use SQL to:
- Extract budget and actual data
- Perform variance calculations
- Aggregate results across periods or departments

Assume the following simplified database tables:

- `budget_financials`
- `actual_financials`

Both tables contain monthly revenue and expense data.


**budget_financials**

| month | revenue_budget | expense_budget |
|------|----------------|----------------|
| Jan  | 100000         | 60000          |
| Feb  | 120000         | 70000          |
| Mar  | 130000         | 75000          |


**actual_financials**

| month | revenue_actual | expense_actual |
|------|----------------|----------------|
| Jan  | 95000          | 62000          |
| Feb  | 125000         | 68000          |
| Mar  | 128000         | 77000          |


**Joining Budget and Actual Data**

To perform variance analysis, budget and actual data
must be joined on a common dimension such as month.

```sql
SELECT
    b.month,
    b.revenue_budget,
    a.revenue_actual,
    b.expense_budget,
    a.expense_actual
FROM budget_financials b
JOIN actual_financials a
    ON b.month = a.month;
```

**Calculating Variances in SQL**

Variance calculations can be performed directly within SQL,
allowing analysts to generate analysis-ready datasets.

```sql
SELECT
    b.month,
    b.revenue_budget,
    a.revenue_actual,
    a.revenue_actual - b.revenue_budget AS revenue_variance,
    ((a.revenue_actual - b.revenue_budget) / b.revenue_budget) * 100
        AS revenue_variance_pct,
    b.expense_budget,
    a.expense_actual,
    a.expense_actual - b.expense_budget AS expense_variance,
    ((a.expense_actual - b.expense_budget) / b.expense_budget) * 100
        AS expense_variance_pct
FROM budget_financials b
JOIN actual_financials a
    ON b.month = a.month;
```

**Interpreting SQL-Based Analysis**

Performing variance analysis in SQL ensures that:
- Calculations are consistent across reports
- Large datasets can be processed efficiently
- Downstream tools receive clean, structured data

**SQL in Real FP&A Workflows**

In practice, SQL-based variance analysis often feeds:
- Business intelligence dashboards
- Monthly management reports
- Automated reporting pipelines

Strong SQL skills enable FP&A analysts
to work directly with source data.


## 8. FP&A Reporting and Insights


**What is FP&A Reporting?**

FP&A reporting is the process of presenting
financial performance data and insights
to management in a clear, actionable format.

Reports summarize:
- Budget vs Actual performance
- Key variances
- Trends over time
- Insights and recommendations

**Common FP&A Report Formats**

Reports can be delivered in multiple formats:

1. **Tabular Reports**: Structured tables showing budgets, actuals, and variances
2. **Charts and Graphs**: Line charts, bar charts, waterfall charts for trends
3. **Dashboards**: Interactive tools that allow drill-downs by period, department, or product

**Key Principles for Reporting**

Effective FP&A reporting follows these principles:

- **Clarity**: The report should be easy to read and understand
- **Actionable Insights**: Highlight deviations and trends
- **Consistency**: Maintain standard formatting across periods
- **Accuracy**: Data must reconcile with source systems

Example: From Variance Analysis to Insight
Consider the variance analysis we performed earlier:

| Month  | Revenue_Variance | Expense_Variance |
|--------|----------------|----------------|
| Jan    | -5,000         | 2,000          |
| Feb    | 5,000          | -2,000         |
| Mar    | -2,000         | 2,000          |

From this table, an analyst can draw insights:
- January: Revenue underperformed; expenses overran → take corrective action
- February: Revenue exceeded plan; expenses were lower → favorable month
- March: Minor revenue shortfall; expenses overrun → monitor cost drivers


**Communicating Insights**

FP&A analysts should always include **interpretation** and **recommendations**:

- What happened?
- Why did it happen?
- What should management do next?

This transforms numbers into **business decisions**, which is the ultimate goal of FP&A.

Reporting Tools

Common tools used in FP&A reporting:

- **Excel / Google Sheets**: Quick, flexible reporting
- **Power BI / Tableau**: Interactive dashboards for management
- **Python / SQL**: Automate data preparation and calculations


A great FP&A analyst does not just present numbers.

1. Numbers → Metrics
