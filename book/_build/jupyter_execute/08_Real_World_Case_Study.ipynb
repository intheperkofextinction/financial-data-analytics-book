## 8.2 Financial Data Cleaning

Before performing any financial analysis, analysts must ensure that the underlying data is accurate, consistent, and analysis-ready.

In real-world FP&A environments, data often comes from multiple systems such as ERP tools, spreadsheets, or data warehouses.

In real-world financial analysis, raw data is rarely analysis-ready.

Common issues include:
- Missing values

- Incorrect data types
- Duplicate records
- Mismatches between budget and actual data

If these issues are not addressed,
financial metrics and FP&A insights can become misleading.

## 8.3 Financial Metric Analysis

With clean and validated data, we can now analyze the financial performance of the business.

In this section, we calculate core financial metrics commonly used in FP&A to evaluate:
- Profitability
- Cost structure
- Operating efficiency
- Trend consistency over time

The goal is not just to compute numbers, but to understand what they reveal about business performance.


### Key Metrics Used in This Analysis

From an FP&A perspective, we focus on metrics that directly influence management decisions:

- **Gross Profit**
- **Gross Margin**
- **Operating Profit**
- **Operating Margin**
- **Cost Ratios (COGS %, Opex %)**
- **Trend Analysis (YoY / directional movement)**

These metrics help answer questions such as:
- Is profitability improving?
- Are costs scaling efficiently with revenue?
- Is performance aligned with budget expectations?


### Profitability Metrics

Profitability metrics show how effectively the company converts revenue into profit.

We start by calculating:
- Gross Profit
- Operating Profit


```python
# Profit calculations
df["Gross_Profit"] = df["Revenue"] - df["COGS"]
df["Operating_Profit"] = df["Gross_Profit"] - df["Operating_Expenses"]

df[["Date", "Revenue", "Gross_Profit", "Operating_Profit"]].head()
```

### Margin Analysis

Margins normalize profit relative to revenue, allowing analysts to:
- Compare performance across periods
- Identify efficiency improvements or deterioration

We calculate:
- Gross Margin
- Operating Margin


```python
# Margin calculations
df["Gross_Margin"] = df["Gross_Profit"] / df["Revenue"]
df["Operating_Margin"] = df["Operating_Profit"] / df["Revenue"]

df[["Date", "Gross_Margin", "Operating_Margin"]].head()
```

### Cost Structure Analysis

Understanding how costs behave relative to revenue is critical in FP&A.

We analyze:
- COGS as a percentage of revenue
- Operating expenses as a percentage of revenue

This helps identify operating leverage and cost discipline.


```python
# Cost ratios
df["COGS_Ratio"] = df["COGS"] / df["Revenue"]
df["Opex_Ratio"] = df["Operating_Expenses"] / df["Revenue"]

df[["Date", "COGS_Ratio", "Opex_Ratio"]].head()
```

### Trend Analysis

FP&A rarely looks at a single period in isolation.

Instead, analysts focus on:
- Directional improvement or deterioration
- Consistency of performance over time
- Early warning signals

We now observe how key metrics evolve month over month.


```python
# Select key metrics for trend review
trend_df = df[[
    "Date",
    "Revenue",
    "Gross_Margin",
    "Operating_Margin",
    "COGS_Ratio",
    "Opex_Ratio"
]]

trend_df.tail()
```

### FP&A Interpretation of Results

Key observations from the metric analysis:

- Revenue shows a steady upward trend, indicating business growth
- Gross margins remain relatively stable, suggesting controlled production costs
- Operating margins improve gradually, reflecting better operating leverage
- Operating expense ratios remain disciplined as revenue scales

From an FP&A standpoint, this indicates:
- Sustainable growth
- Effective cost management
- Improving profitability quality, not just top-line expansion


### Business Implications

Based on these metrics, FP&A teams would likely:
- Support continued investment in growth initiatives
- Monitor operating expenses closely to preserve margin gains
- Use current performance as a baseline for forecasting and scenario analysis

These insights form the foundation for budget reviews, reforecasting, and strategic planning.


## 8.4 FP&A Insights & Variance Interpretation

Financial metrics alone do not drive decisions.

FP&A teams focus on **variance analysis** — understanding:
- What deviated from plan
- Why it happened
- Whether it is controllable
- What action management should take

In this section, we compare **Actual vs Budgeted performance** and interpret the results like an FP&A analyst.


### Why Variance Analysis Matters

Variance analysis is the core responsibility of FP&A.

Management expects answers to:
- Why did revenue miss or beat budget?
- Are cost overruns structural or temporary?
- Which variances require action vs monitoring?

Good FP&A analysis separates **signal from noise** and converts numbers into decisions.


```python
# Revenue variance
df["Revenue_Variance"] = df["Revenue"] - df["Budget_Revenue"]
df["Revenue_Variance_Pct"] = df["Revenue_Variance"] / df["Budget_Revenue"]

df[["Date", "Revenue", "Budget_Revenue", "Revenue_Variance", "Revenue_Variance_Pct"]].head()
```

### Revenue Variance Interpretation

Revenue variance measures demand-side performance.

Positive variance may indicate:
- Stronger-than-expected sales volume
- Pricing improvements
- Favorable market conditions

Negative variance may indicate:
- Demand softness
- Competitive pressure
- Execution gaps

FP&A focuses on **trend consistency**, not one-off months.


```python
# COGS variance
df["COGS_Variance"] = df["COGS"] - df["Budget_COGS"]
df["COGS_Variance_Pct"] = df["COGS_Variance"] / df["Budget_COGS"]

df[["Date", "COGS", "Budget_COGS", "COGS_Variance", "COGS_Variance_Pct"]].head()
```

### COGS Variance Interpretation

COGS variance reflects production and procurement efficiency.

Unfavorable COGS variance may suggest:
- Input cost inflation
- Supplier pricing pressure
- Manufacturing inefficiencies

Favorable variance may indicate:
- Cost optimization initiatives
- Improved supplier terms
- Operational scale benefits

FP&A evaluates whether these variances are **temporary or structural**.


```python
# Operating expense variance
df["Opex_Variance"] = df["Operating_Expenses"] - df["Budget_Opex"]
df["Opex_Variance_Pct"] = df["Opex_Variance"] / df["Budget_Opex"]

df[["Date", "Operating_Expenses", "Budget_Opex", "Opex_Variance", "Opex_Variance_Pct"]].head()
```

### Operating Expense Variance Interpretation

Operating expenses are usually more controllable than revenue.

Unfavorable opex variance may result from:
- Hiring acceleration
- Marketing spend increases
- One-time operational costs

FP&A distinguishes:
- Strategic overspend (intentional)
- Execution slippage (unplanned)

This distinction drives very different management responses.


```python
# Budget vs Actual operating profit
df["Budget_Operating_Profit"] = (
    df["Budget_Revenue"] - df["Budget_COGS"] - df["Budget_Opex"]
)

df["Operating_Profit_Variance"] = (
    df["Operating_Profit"] - df["Budget_Operating_Profit"]
)

df[["Date", "Operating_Profit", "Budget_Operating_Profit", "Operating_Profit_Variance"]].head()
```

### Integrated FP&A Insight

Looking at variances together provides a complete picture:

- Revenue is consistently meeting or exceeding budget
- Cost overruns are present but controlled
- Operating profit remains resilient due to scale benefits

This indicates:
- Healthy operating leverage
- Manageable cost pressures
- Strong execution against plan


### FP&A-Driven Management Actions

Based on this analysis, FP&A would recommend:

- Maintain current growth investments
- Monitor COGS closely for inflation risks
- Tighten discretionary opex controls where variance persists
- Use current performance as input for rolling forecasts

FP&A does not just report variance — it **guides decisions**.


### Analyst Takeaway

Variance analysis is not about finding mistakes.

It is about:
- Understanding business dynamics
- Supporting leadership decisions
- Balancing growth and control

Strong FP&A analysts combine numerical rigor with business judgment.
